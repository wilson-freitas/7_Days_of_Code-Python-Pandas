## - Concatenando os dados para trabalharmos em um dataframe único

```python
# Criando uma lista com os diferentes dataframes, podemos utilizar o método .concat do pandas
emprestimos_biblioteca = pd.concat ([dados_2010_1, dados_2010_2 ,dados_2011_1, dados_2011_2, dados_2012_1, dados_2012_2, dados_2013_1,
                                dados_2013_2, dados_2014_1, dados_2014_2, dados_2015_1, dados_2015_2, dados_2016_1, dados_2016_2,
                                dados_2017_1, dados_2017_2, dados_2018_1, dados_2018_2, dados_2019_1, dados_2019_2, dados_2020_1])

emprestimos_biblioteca # Mostrando o dataframe concatenado
```

No output acima, podemos perceber que a unificação deu certo, temos dados desde 2010 até 2021:

## - Verificando duplicatas:

```python
# O método value_counts() nos permite verificar a quantidade de vezes que uma linha aparece no nosso dataframe
emprestimos_biblioteca.value_counts()
```

Na primeira linha, o número 7 ao final dos valores representa a quantidade de vezes que a linha de valores apareceu no nosso conjunto de dados

## - Excluindo duplicatas:

```python
emprestimos_biblioteca = emprestimos_biblioteca.drop_duplicates()
```

Agora, utilizamos o value_counts() para verificar se as duplicatas foram removidas com sucesso:

```python
emprestimos_biblioteca.value_counts()
```

Vemos que cada linha só aparece 1 vez no conjunto (numero 1 ao final de cada linha), sendo assim, nosso conjunto não tem duplicatas

## - Importando dados dos exemplares de um arquivo parquet

Arquivos parquet servem para armazenar grandes volumes de dados de maneira eficiente, o método pd.read_parquet() do pandas lê esse tipo de arquivo e armazena-o em um dataframe do panda 

```python
dados_exemplares = pd.read_parquet('https://github.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/raw/main/Dia_1-Importando_dados/Datasets/dados_exemplares.parquet')
dados_exemplares
```

## - Verificando duplicatas nos dados dos exemplares

```python
dados_exemplares.value_counts()
```

Duplicatas verificadas, podemos proseguir para a junção dos dados_exeplares com os emprestimos_bibliotecas

## - Realizando a junção dos dados

```python
emprestimos_completo = emprestimos_biblioteca.merge(dados_exemplares)
emprestimos_completo
```

O método merge realiza a junção dos dados com base em colunas correspondentes, nesse caso, apenas a coluna codigo_barras
